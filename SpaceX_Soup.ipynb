# Parsing data using BeautifulSoup

  In this module the data will be obtain from wikipedia from the third table named 'List of Falcon 9 and Falcon Heavy launches', obtained from the next link: https://en.wikipedia.org/wiki/List_of_Falcon_9_and_Falcon_Heavy_launches
  

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/List_of_Falcon_9_and_Falcon_Heavy_launches_(2010%E2%80%932019)'
response= requests.get(url)
response
```

```python
soup= BeautifulSoup(response.text,'html.parser')
soup.title
```

```python
lunch_table= soup.find_all('table')[5]
lunch_table
```

```python
soup.find_all('th')
```

```python
table_columns = []
#flights=[]
for row in lunch_table.find_all('th'):
    if row.a:
        row.a.extract()
    if row.b:
        row.b.extract()
    if row.sup:
        row.sup.extract()
        
    values = row.get_text(strip=True, separator=" ")
    if not values.isdigit() and values !='' and values!='FH 4':
        table_columns.append(values)
   # if values.isdigit() or values =='FH 4':
   #     flights.append(values)
        
 
print(table_columns)
```

```python
Launch_dict= dict.fromkeys(table_columns)

del Launch_dict['Date and time ( )']

#general variables

Launch_dict['Flight No.']=[]
Launch_dict['Launch site']=[]
Launch_dict['Payload']=[]
Launch_dict['Payload mass']=[]
Launch_dict['Orbit']=[]
Launch_dict['Customer']=[]
Launch_dict['Launch outcome']=[]
#added columns
Launch_dict['Version Booster']=[]
Launch_dict['Booster landing']=[]
Launch_dict['Date']=[]
Launch_dict['Time']=[]

print(Launch_dict)
```

## Understanding the basic coding (i)
 it is possible iterate over a tag with the beautifulSoup method using the commmand <b> object.strings </b> in a 'for' loop , so in these case as the data was tranform into a string value, the method <b>split(' ')</b> was used.
 This split() will give you a list, that can be transform into a String value using for example:

 
<strong> List_to_String= ' '.join(new_list[0:3])<strong/>
 
 In this way it is possible obtain data of date and time from the <b>td<b/>.

 ### iteration trough rows (td)
  As the table has sub-rows in some cases, given an explanation of the launching or another details related with other columns. Due to this, to avoid considering this data at the moment of extracting information,<b> there is a filter that consider just those rows with a total count of data/column superior to 5</b> , this same process is done with the headers in the coding made by IBM.
  

```python

#fill the flight number list with the list created before
flights=[]
#for row in flights:
#    Launch_dict['Flight No.'].append(row)
for i,lunch_table in enumerate(soup.find_all('table')):
    if i<7:
        for row in lunch_table.find_all('th',{"scope": "row"}):
                
            values = row.get_text(strip=True, separator=" ")
              
            #if values or values in ['FH 1','FH 2','FH 3','N/A']:
            flights.append(values)
            for row in flights:
                Launch_dict['Flight No.'].append(row)
                flights=[]
                
        
        for i,row in enumerate(lunch_table.find_all('tr')):
            values=row.find_all('td')
            if len(list(row.find_all('td')))>7 and values!='':
                if len(values)>1:
                    booster=[]
                    for i,cell in enumerate(values[1].strings):
                        if i%2==0:
                            booster.append(cell)
                            bv=' '.join(booster[:2])
            
                    Launch_dict['Version Booster'].append(bv)
                    booster=[]
                if len(values)>2:
                    Launch_dict['Launch site'].append(values[2].get_text(strip=True).split(',')[0])
                if len(values)>3:
                    booster=[]
                    for cell in values[3].strings:
                        booster.append(cell)
                    Launch_dict['Payload'].append(booster[0])
                    booster=[]
                if len(values)>4:
                    booster=[]
                    for cell in values[4].strings:
                        booster.append(cell)
                    mass_kg=booster[0].split(' ')[0][:-3].replace('~','').replace(',','')
                    try: 
                        Launch_dict['Payload mass'].append(int(mass_kg))
                        booster=[]
                    except:
                        Launch_dict['Payload mass'].append(int('0'))
                        booster=[]
                if len(values)>5:
                    booster=[]
                    for cell in values[5].strings:
                        booster.append(cell)
                    if booster:
                        Launch_dict['Orbit'].append(booster[0])
                    else:
                        Launch_dict['Orbit'].append('Unknown')
                    booster=[]
                if len(values)>6:
                    booster=[]
                    for cell in values[6].strings:
                        booster.append(cell.strip())
                    Launch_dict['Customer'].append(booster[0])
                    booster=[]
                if len(values)>7:
                    booster=[]
                    for cell in values[7].strings:
                        booster.append(cell.strip())
                    Launch_dict['Launch outcome'].append(booster[0])
                    booster=[]
                if len(values)>8:
                    booster=[]
                    for cell in values[8].strings:
                        booster.append(cell.strip())
                    Outcome=' '.join(booster[:2])
                    Launch_dict['Booster landing'].append(Outcome)
                    booster=[]
                if len(values)>0:
                    data=[]
                    data_value= values[0].get_text(strip=(','),separator=' ').split(' ')
                    for string in data_value:
                        data.append(string.strip())
                    data_str=' '.join(data[0:3])
                    time_str=' '.join(data[3:4])
                    Launch_dict['Date'].append(data_str)
                    Launch_dict['Time'].append(time_str)
                    data=[]
    
Launch_dict
```
